```python
# OVERLAY SMALL IMAGE  ON TOP OF A LARGER IMAGE (BO BLENDING)
# NUMPY REASSIGNMENT
```

```python
img1 = cv.imread("puppy.jpg")
img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)

img2 = cv.imread("watermark.png")
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
```

```python
img2 = cv.resize(img2, (600,600))
```

```python
plt.imshow(img2)
```

```python
large_img = img1 
small_img = img2
```

```python
x_offset = 100
y_offset = 100
```

```python
x_end =  x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```
